# Dog Breed Classifier

If you have a cute puppy, but you don't know where it came from, here's your chance for clarity.

Upload a picture of your puppy and get a guess of what breed it is. 🐶

```
from fastai.vision.all import *
from fastai.vision.widgets import *
```

```
path = Path()
learn_inf = load_learner(path / 'breedclassifier.pkl')

# creating all our widget elements
btn_upload = widgets.FileUpload()
out_pl = widgets.Output()
lbl_pred = widgets.Label()
lbl_test = widgets.Label()
lbl_test.value = "Hello!"
btn_run = widgets.Button(description='Classify')
```

```
def on_click_classify(change):
    lbl_pred.value = ''
    img = PILImage.create(btn_upload.data[-1])
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(128,128))
    pred,pred_idx,probs = learn_inf.predict(img)
    lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'
```

```
btn_run.on_click(on_click_classify)
```

```
display(VBox([widgets.Label('Select your dog! Then click "Classify"'), 
              btn_upload, btn_run, out_pl, lbl_test, lbl_pred, lbl_test]))
```
